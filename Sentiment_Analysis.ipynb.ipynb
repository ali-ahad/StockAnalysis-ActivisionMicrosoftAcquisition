```python
import pandas as pd
from transformers import AutoModelForSequenceClassification, AutoTokenizer, AutoConfig
import numpy as np
from scipy.special import softmax

# 定义预处理函数
def preprocess(text, max_length=512):
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    # 将文本合并并截断
    return " ".join(new_text)[:max_length]


# 加载模型和分词器
MODEL = "cardiffnlp/twitter-roberta-base-sentiment-latest"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)

# 函数：对单个文本进行情感分析
def sentiment_analysis(text):
    processed_text = preprocess(text)
    encoded_input = tokenizer(processed_text, return_tensors='pt')
    output = model(**encoded_input)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    return scores

# 为DataFrame添加新列
comments_df['Sentiment Scores'] = comments_df['Comment Body'].apply(sentiment_analysis)

# 将得分转换为类别
config = AutoConfig.from_pretrained(MODEL)
def get_sentiment_category(scores):
    ranking = np.argsort(scores)
    ranking = ranking[::-1]
    return config.id2label[ranking[0]]

comments_df['Sentiment Category'] = comments_df['Sentiment Scores'].apply(get_sentiment_category)

# 展示更新后的DataFrame
print(comments_df)
```

```python
# 将日期列转换为 datetime 对象
comments_df['Comment Time'] = pd.to_datetime(comments_df['Comment Time'])

# 将日期格式化为 'YYYY-MM-DD' 格式
comments_df['Comment Time'] = comments_df['Comment Time'].dt.strftime('%Y-%m-%d')

# 将列名从 'Comment Time' 改为 'date'
comments_df.rename(columns={'Comment Time': 'date'}, inplace=True)
comments_df.rename(columns={'Comment Body': 'comment'}, inplace=True)
comments_df.rename(columns={'Sentiment Scores': 'score'}, inplace=True)
comments_df.rename(columns={'Sentiment Category': 'category'}, inplace=True)
```

```python
import pandas as pd

# 假设 comments_df 是包含 'date' 和 'category' 列的DataFrame

# 按日期和情感类别分组，计算每组的数量
grouped = comments_df.groupby(['date', 'category']).size().reset_index(name='count')

# 创建pivot table，其中行是日期，列是情感类别，值是计数
pivot_table = grouped.pivot_table(index='date', columns='category', values='count', fill_value=0)

# 确保所有情感类别的列都存在，即使某些类别在某天没有评论
pivot_table = pivot_table.reindex(columns=['positive', 'neutral', 'negative'], fill_value=0)

# 将 'date' 索引转换为列
pivot_table.reset_index(inplace=True)
```

```python
pivot_table['date'] = pd.to_datetime(pivot_table['date'])
pivot_table = pivot_table[pivot_table['date'].dt.year > 2021]
```
